```python
from PIL import Image
from PIL import ImageFilter
```

```python
for i in range(1,6):
    class_num = '\\Class ' + str(i)
    for j in range(1,36):
        file = '\\a' + str(j)
        temp = path + class_num + file + '.jpg'
        
        im = Image.open(temp)
        
        new_path= path + '\\blurred\\' + class_num + file + '.jpg'
        blurred_image = im.filter(ImageFilter.GaussianBlur(radius=3))
        blurred_image.save(new_path)
```

```python
test_path = 'C:\\Users\\RQ\\Downloads\\'

im = Image.open(test_path + 'test2.jpg')
blurred_image = im.filter(ImageFilter.GaussianBlur(radius=4))
blurred_image.save(test_path + 'blur_test2.jpg')
```
